### _Why do we need to do Sampling ?_
Is there a way to pick a subset of the data and analyze that – and that can be a good representation of the entire dataset? Obtaining the entire set of data of interest is too expensive or time-consuming. Also, it’s time-consuming and expensive to process all the data that is collected. So, using a sampling algorithm can reduce the data size where a better, but the more expensive algorithm can be used.

### _What is Sampling ?_

Sampling is a method that allows us to get information about the population based on the statistics from a subset of the population (sample), without having to investigate every individual.

### _What are the different types of Sampling Techniques ?_

There are many different types of Sampling techniques. For the sake of simplicity, i am only going to discuss three different types of Sampling Techniques: <br>
* Simple Random Sampling
* Stratified Random Sampling
* Cluster Random Sampling

PS.  *Note*: I will be using Bank Marketing data which can be found [here](https://archive.ics.uci.edu/ml/datasets/Bank+Marketing).

```python
import pandas as pd
import numpy as np
data=pd.read_csv("bank.csv",delimiter=";")
data.head()
```

### Simple Random Sampling

Every element has an equal chance of getting selected to be the part sample. It is used when we don’t have any kind of prior information about the target population. <br>
E.g. Random selection of 9 people from an MNC of 23 employees. Each student has equal chance of getting selected. Here probability of selection is 1/23

```python
# loading image
from IPython.display import Image
Image(filename="people.jpg")
```

```python
data.shape[0]
```

```python
data.sample(frac=0.70, replace=True)
```

Note: 'replace'  parameter in the above example of Simple random sampling enables us to sample data with or without replacement having a default of False.

You can see that the number of observations now are 3166 which is almost **70%**

### Stratified Random Sampling

This technique divides the elements of the population into small subgroups (strata) based on the similarity in such a way that the elements within the group are homogeneous and heterogeneous among the other subgroups formed. And then the elements are randomly selected from each of these strata.We use this type of sampling when we want representation from all the subgroups of the population. However, stratified sampling requires proper knowledge of the characteristics of the population.

```python
# loading image
from IPython.display import Image
Image(filename="stratified.png")
```

In the above example, you can see that the people have been grouped based on the shirt colours. 

```python
from sklearn.model_selection import train_test_split

stratified_sample, _ = train_test_split(data, test_size=0.50, stratify=data['marital'])
stratified_sample
```

In this example, i have stratified data on Marital column 

### Cluster Sampling

In a clustered sample, we use the subgroups of the population as the sampling unit rather than individuals. The population is divided into subgroups, known as clusters, and a whole cluster is randomly selected to be included in the study.

```python
# loading image
from IPython.display import Image
Image(filename="cluster.jpg")
```

The main aim of cluster sampling can be specified as cost reduction and increasing the levels of efficiency of sampling. This specific technique can also be applied in integration with multi-stage sampling. A major difference between cluster and stratified sampling relates to the fact that in cluster sampling a cluster is perceived as a sampling unit, whereas in stratified sampling only specific elements of strata are accepted as sampling unit.

### Systematic Random Sampling

Systematic sampling is a type of probability sampling method in which sample members from a larger population are selected according to a random starting point but with a fixed, periodic interval. This interval, called the sampling interval, is calculated by dividing the population size by the desired sample size.

Within systematic sampling, as with other sampling methods, a target population must be selected prior to selecting participants. A population can be identified based on any number of desired characteristics that suit the purpose of the study being conducted. Some selection criteria may include age, gender, race, location, education level and/or profession.
